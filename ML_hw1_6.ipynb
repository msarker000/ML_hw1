6A. Assume there are 250 class A points, and 250 class B points. Use
matplot lib to plot these points. These clusters should be very well
separated and any classifier should do very well on them. Please plot
the blobs with class A in blue and class B points as in red. Not that
you have overplotting so you cannot see that your blobs are much
denser in the middle then at the edges. Decrease the size of the dots
and also use a small transparency alpha << 1 untill you can clearly
see dots in the core as been denser. Always beware of overplotting

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline

random_state=42

def generate_sample_data(n_samples, cluster_std, center_box):
    X, y = make_blobs(n_samples=n_samples, n_features=2, cluster_std=cluster_std, center_box=center_box, shuffle=True, random_state=random_state)
    return X, y

#generate  data from Class A and plot using make_blobs
X, y = generate_sample_data(n_samples=250, cluster_std=0.125, center_box=(-1,-1) )
plt.scatter(X[:, 0], X[:, 1], c='b', alpha=0.1, s=10)

#generate data from Class B and plot
X, y = generate_sample_data(n_samples=250, cluster_std=0.125, center_box=(1,1) )
plt.scatter(X[:, 0], X[:, 1], c= 'r', alpha=0.1, s=10)
 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(['Class A','Class B'], loc='upper left')

plt.show()
```

6D. For our last example we want to create data which is clearly separable
but not linearly separaable. Use sklearn.datasets.make_circles to
make two circles in 2D each with 500 points. Unfortuately we don’t
want a little scircle in the middle but a normally disturbed blob.
Use fancy indexing to isolate data on the outer circle only (class

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=100, noise=0.05)
colors = {0:'red', 1:'blue'}

plt.scatter(X[y==0][:, 0], X[y==0][:, 1], c=colors[0])
plt.scatter(X[y==1][:, 0], X[y==1][:, 1], c=colors[1])

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(['0','1'], loc='upper left')
plt.show()
```
